```python
tkl.seed(123)

amp = 5
SCOPES = [0.1, 1, 10, 100, 1000]

plt.figure(figsize=(20, 4))
for i in range(5):
    kernel = tkl.GaussianKernel(SCOPES[i])
    b2 = bati1.noise(amp, kernel, cycle=True, force=True)
    if not ((b2.getX()[-1] == b2.getX()[0]) and (b2.getY()[-1] == b2.getY()[0])):
        b2.addObs(b2[0])
    
    ax = plt.subplot2grid((1,5), (0,i))
    ax.plot(bati1.getX(), bati1.getY(), color="royalblue", linestyle='--')
    ax.plot(b2.getX(), b2.getY(), color="limegreen", linestyle='-')
    ax.set_xticks([]) 
    ax.set_yticks([]) 
```
