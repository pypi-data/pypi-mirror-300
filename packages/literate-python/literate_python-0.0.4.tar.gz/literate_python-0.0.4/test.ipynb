# Prepare a server for literate python

```python
import os
from lpy.server import run_server as run_lpy_server
from threading import Thread

os.environ['LITERATE_PYTHON_HOST'] = '127.0.0.1'
os.environ['LITERATE_PYTHON_PORT'] = '7331'
lpy_server_thread = Thread(target=run_lpy_server)
lpy_server_thread.start()
```

```python
from lpy.server import run_websockets_server
import os
from threading import Thread

os.environ["LITERATE_PYTHON_PORT"] = '7332'
lpy_websocket_server_thread = Thread(target=run_websockets_server)
lpy_websocket_server_thread.start()
```
