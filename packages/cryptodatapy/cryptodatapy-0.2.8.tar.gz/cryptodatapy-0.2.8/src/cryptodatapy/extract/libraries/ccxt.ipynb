```python
import pandas as pd
from cryptodatapy.extract.libraries.ccxt_api import CCXT
from cryptodatapy.extract.datarequest import DataRequest
from cryptodatapy.transform.convertparams import ConvertParams
from cryptodatapy.extract.getdata import GetData
```

```python
cx = CCXT()
```

```python
# get all Binance perp futures tickers
data_req = DataRequest(source='ccxt')
perp_tickers = GetData(data_req).get_meta(method='get_markets_info', 
                                          quote_ccy='USDT',
                                          mkt_type='perpetual_future',
                                          exch='binanceusdm', 
                                          as_list=True)
```

```python
perp_tickers
```

```python
start = pd.Timestamp.now()
```

```python
data_req = DataRequest(source='ccxt',
                       exch = 'binanceusdm',
                       source_markets=perp_tickers,
                       fields=['open', 'high', 'low', 'close', 'volume'],
                       freq='1h',
                       quote_ccy=None,
                       mkt_type='perpetual_future',
                       start_date=None,
                       end_date=None,
                       trials=3,
                       pause=0.5
                      ) 
```

```python
df = await cx.get_data(data_req)
```

```python
pd.Timestamp.now() - start
```

```python
df
```

```python
df.to_parquet('s3://factorlab-data/binance_perps_ohlcv_hourly.parquet')
```
