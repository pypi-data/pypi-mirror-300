## Query: Base units -> use filter for conversionMultiplier (1.0)

```python
# query has multiple units
# -> distinct works in combination with other attributes 
# -> delete ?versonNo, ucumCode

PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX owl: <http://www.w3.org/2002/07/owl#>
PREFIX qudt: <http://qudt.org/schema/qudt/>
PREFIX quantitykind: <http://qudt.org/vocab/quantitykind/>
PREFIX unit: <http://qudt.org/vocab/unit/>

SELECT DISTINCT ?unit ?label ?quantity ?iec61360Code ?ucumCode ?conversionMultiplier ?versionNo
{
  ?unit a qudt:Unit ; 
    	rdfs:label 					?label ;
    	qudt:hasQuantityKind 		?quantity ;
    	qudt:iec61360Code 			?iec61360Code ;
    	qudt:ucumCode 				?ucumCode ;
    	qudt:conversionMultiplier	?conversionMultiplier;
    	rdfs:isDefinedBy ?ontology .
  filter (lang(?label)="en")
  filter ((?conversionMultiplier)=1.0)
  ?ontology a owl:Ontology ; owl:versionIRI ?version .
  bind ( REPLACE(STR(?version), "^.*org/([^/]+).*$", "$1") as ?versionNo )
} limit 10000
```

```python
# query has multiple units
# -> distinct works in combination with other attributes 
# -> delete all but unit & conversionMultiplier

PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX owl: <http://www.w3.org/2002/07/owl#>
PREFIX qudt: <http://qudt.org/schema/qudt/>
PREFIX quantitykind: <http://qudt.org/vocab/quantitykind/>
PREFIX unit: <http://qudt.org/vocab/unit/>

SELECT DISTINCT ?unit ?label ?quantity ?iec61360Code ?conversionMultiplier 
{
  ?unit a qudt:Unit ; 
    	rdfs:label 					?label ;
    	qudt:hasQuantityKind 		?quantity ;
    	qudt:iec61360Code 			?iec61360Code ;
    	qudt:conversionMultiplier	?conversionMultiplier;
  filter (lang(?label)="en")
  filter ((?conversionMultiplier)=1.0)
} limit 10000
```

```python
# query all units without ?symbol attribute
# count: 445

PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX owl: <http://www.w3.org/2002/07/owl#>
PREFIX qudt: <http://qudt.org/schema/qudt/>


SELECT ?unitIRI ?unitName ?conversionMultiplier ?quantity 
WHERE {
  	?unitIRI  rdf:type qudt:Unit ;
    	qudt:conversionMultiplier	?conversionMultiplier ;
		qudt:hasQuantityKind 		?quantity ;

  	FILTER ((?conversionMultiplier)=1.0)
  	BIND ( REPLACE(STR(?unitIRI), "^.*unit/([^/]+).*$", "$1") as ?unitName)
} 
ORDER BY ASC (?quantity)
LIMIT 10000
```

```python
# figure out how many and which units don't have a dimension vector
# count: 7

PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX owl: <http://www.w3.org/2002/07/owl#>
PREFIX qudt: <http://qudt.org/schema/qudt/>

SELECT ?quantityIRI ?quantityName ?unitIRI ?unitName ?conversionMultiplier ?isBaseUnit ?hasDimensionVector
WHERE {
  	?unitIRI  rdf:type qudt:Unit ;
    	qudt:conversionMultiplier	?conversionMultiplier ;
		qudt:hasQuantityKind 		?quantityIRI ;
  	FILTER ((?conversionMultiplier)!=0.0)
  	BIND ( REPLACE(STR(?unitIRI), "^.*unit/([^/]+).*$", "$1") AS ?unitName)
    BIND ( REPLACE(STR(?quantityIRI), "^.*quantitykind/([^/]+).*$", "$1") AS ?quantityName)
  	BIND ( IF(?conversionMultiplier = 1.0, True, False)  AS ?isBaseUnit)
	FILTER NOT EXISTS {?unitIRI qudt:hasDimensionVector		?hasDimensionVector}
} 
ORDER BY ASC (?unitName)

LIMIT 10000

```
